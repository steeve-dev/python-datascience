```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('healthexp')
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe().transpose()
```

```python
data['Country'].value_counts()
```

On compare l'évolution des dépenses de santé dans chaque pays

```python
sns.lineplot(x="Year", y="Spending_USD", hue="Country", data=data)
```

On compare l'evolution de l'espérance de vie de chaque pays

```python
sns.lineplot(x="Year", y="Life_Expectancy", hue="Country", data=data)
```

Le prochain graphique sert à représenter l'évolution des dépenses de santé et l'évolution de l'espèrance de vie de chaque pays

```python
sns.scatterplot(data=data, x='Life_Expectancy', y='Spending_USD', hue='Country')
```

```python
correlation = data.corr(numeric_only=True)
correlation
```

```python
correlation['Spending_USD']
```

```python
plt.figure(figsize=(8,4))

sns.heatmap(correlation,annot=True,cmap='crest')
```
